1)	Aplique o filtro de Média a imagem abaixo para diferentes tamanhos de máscaras.
- 3x3; 7x7; 3x3 aplicado três vezes sequenciais.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python


img1 = cv2.imread("sta2.png", cv2.IMREAD_GRAYSCALE)

mask_3 = np.ones((3,3), np.float32)/9
mask_5 = np.ones((5,5), np.float32)/25
mask_7 = np.ones((7,7), np.float32)/49

def mean_filter(image, kernel):
  return cv2.filter2D(image, -1, kernel)

image_3 = np.copy(img1)
image_3 = mean_filter(image_3, mask_3)
image_3 = mean_filter(image_3, mask_3)
image_3 = mean_filter(image_3, mask_3)

image_5 = np.copy(img1)
image_5 = mean_filter(image_5, mask_5)
image_5 = mean_filter(image_5, mask_5)
image_5 = mean_filter(image_5, mask_5)

image_7 = np.copy(img1)
image_7 = mean_filter(image_7, mask_7)
image_7 = mean_filter(image_7, mask_7)
image_7 = mean_filter(image_7, mask_7)


plt.imshow(img1, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()
plt.imshow(image_3, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()
plt.imshow(image_5, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()
plt.imshow(image_7, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()
```

2)	Aplique o filtro Gaussiano à imagem abaixo para diferentes valores de desvio padrão (σ) e tamanho de máscara.
- σ = 1.0 (kernel size 5×5).	σ = 2.0 (kernel size 9×9).	σ =  4.0 (kernel size 15×15).

```python
#  code here
img2 = cv2.imread("ben2.png", cv2.IMREAD_GRAYSCALE)

gaussian_1 = np.copy(img2)
gaussian_1 = cv2.GaussianBlur(gaussian_1, (5,5), 1)

gaussian_2 = np.copy(img2)
gaussian_2 = cv2.GaussianBlur(gaussian_2, (9,9), 2)

gaussian_3 = np.copy(img2)
gaussian_3 = cv2.GaussianBlur(gaussian_3, (15,15), 4)

plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()
plt.imshow(gaussian_1, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()
plt.imshow(gaussian_2, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()
plt.imshow(gaussian_3, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()
```

3)	Aplique o filtro de Mediana para a imagem abaixo, que apresenta um ruído *salt and pepper*
-	3×3 filter;	7×7 filter

```python
#  code here
img3 = cv2.imread("sta2noi2.png", cv2.IMREAD_GRAYSCALE)

median_1 = np.copy(img3)
median_1 = cv2.medianBlur(median_1, 3)

median_2 = np.copy(img3)
median_2 = cv2.medianBlur(median_1, 7)

plt.imshow(img3, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()
plt.imshow(median_1, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()
plt.imshow(median_2, cmap='gray', vmin=0, vmax=255)
plt.show(block=False)
plt.close()
```
